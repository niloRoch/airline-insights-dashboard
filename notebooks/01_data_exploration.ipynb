```python
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# üìä 01 - Data Exploration: Airlines Dataset\n",
    "## An√°lise Explorat√≥ria de Dados - Voos Delhi-Mumbai\n",
    "\n",
    "**Objetivo**: Compreender a estrutura, qualidade e caracter√≠sticas principais do dataset de voos.\n",
    "\n",
    "**Autor**: [Seu Nome]  \n",
    "**Data**: $(date +\"%Y-%m-%d\")  \n",
    "**Vers√£o**: 1.0\n",
    "\n",
    "---"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üîß Setup e Importa√ß√µes"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Importa√ß√µes principais\n",
    "import pandas as pd\n",
    "import numpy as np\n",
    "import matplotlib.pyplot as plt\n",
    "import seaborn as sns\n",
    "import plotly.express as px\n",
    "import plotly.graph_objects as go\n",
    "from plotly.subplots import make_subplots\n",
    "import warnings\n",
    "\n",
    "# Configura√ß√µes\n",
    "warnings.filterwarnings('ignore')\n",
    "plt.style.use('seaborn-v0_8')\n",
    "sns.set_palette(\"husl\")\n",
    "pd.set_option('display.max_columns', None)\n",
    "pd.set_option('display.max_rows', 100)\n",
    "\n",
    "# Configura√ß√µes do Plotly\n",
    "import plotly.io as pio\n",
    "pio.templates.default = \"plotly_white\"\n",
    "\n",
    "print(\"üì¶ Todas as bibliotecas carregadas com sucesso!\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üì• Carregamento dos Dados"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Carregar dataset\n",
    "df = pd.read_csv('../data/raw/airlines_flights_data.csv')\n",
    "\n",
    "print(f\"‚úÖ Dataset carregado com sucesso!\")\n",
    "print(f\"üìä Dimens√µes: {df.shape[0]} linhas √ó {df.shape[1]} colunas\")\n",
    "print(f\"üíæ Tamanho em mem√≥ria: {df.memory_usage(deep=True).sum() / 1024:.2f} KB\")\n",
    "\n",
    "# Preview dos dados\n",
    "print(\"\\nüìã Primeiras 5 linhas:\")\n",
    "df.head()"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üîç An√°lise Estrutural dos Dados"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Informa√ß√µes gerais do dataset\n",
    "print(\"üìã INFORMA√á√ïES GERAIS DO DATASET\")\n",
    "print(\"=\" * 50)\n",
    "df.info()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise das colunas\n",
    "print(\"üè∑Ô∏è AN√ÅLISE DAS COLUNAS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "column_analysis = pd.DataFrame({\n",
    "    'Coluna': df.columns,\n",
    "    'Tipo': df.dtypes,\n",
    "    'Valores_√önicos': [df[col].nunique() for col in df.columns],\n",
    "    'Valores_Nulos': [df[col].isnull().sum() for col in df.columns],\n",
    "    'Percentual_Nulos': [f\"{(df[col].isnull().sum() / len(df)) * 100:.1f}%\" for col in df.columns]\n",
    "})\n",
    "\n",
    "column_analysis"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Verificar duplicatas\n",
    "duplicates = df.duplicated().sum()\n",
    "print(f\"üîÑ Linhas duplicadas: {duplicates}\")\n",
    "\n",
    "if duplicates > 0:\n",
    "    print(\"‚ö†Ô∏è Linhas duplicadas encontradas:\")\n",
    "    display(df[df.duplicated(keep=False)].sort_values('flight'))\n",
    "else:\n",
    "    print(\"‚úÖ Nenhuma linha duplicada encontrada\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üìà An√°lise Univariada - Vari√°veis Categ√≥ricas"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise das companhias a√©reas\n",
    "print(\"‚úàÔ∏è DISTRIBUI√á√ÉO POR COMPANHIA A√âREA\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "airline_counts = df['airline'].value_counts()\n",
    "airline_percentage = df['airline'].value_counts(normalize=True) * 100\n",
    "\n",
    "airline_summary = pd.DataFrame({\n",
    "    'Voos': airline_counts,\n",
    "    'Percentual': airline_percentage.round(1)\n",
    "})\n",
    "\n",
    "print(airline_summary)\n",
    "\n",
    "# Visualiza√ß√£o\n",
    "fig = px.pie(values=airline_counts.values, \n",
    "             names=airline_counts.index, \n",
    "             title=\"Distribui√ß√£o de Voos por Companhia A√©rea\",\n",
    "             hole=0.4)\n",
    "fig.update_layout(height=500, showlegend=True)\n",
    "fig.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise dos hor√°rios de partida\n",
    "print(\"üïê DISTRIBUI√á√ÉO POR HOR√ÅRIO DE PARTIDA\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "departure_counts = df['departure_time'].value_counts()\n",
    "departure_percentage = df['departure_time'].value_counts(normalize=True) * 100\n",
    "\n",
    "departure_summary = pd.DataFrame({\n",
    "    'Voos': departure_counts,\n",
    "    'Percentual': departure_percentage.round(1)\n",
    "})\n",
    "\n",
    "print(departure_summary)\n",
    "\n",
    "# Visualiza√ß√£o\n",
    "fig = px.bar(x=departure_counts.index, \n",
    "             y=departure_counts.values,\n",
    "             title=\"Distribui√ß√£o de Voos por Hor√°rio de Partida\",\n",
    "             labels={'x': 'Hor√°rio de Partida', 'y': 'N√∫mero de Voos'})\n",
    "fig.update_layout(height=400, xaxis_tickangle=-45)\n",
    "fig.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise das paradas\n",
    "print(\"üõë DISTRIBUI√á√ÉO POR N√öMERO DE PARADAS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "stops_counts = df['stops'].value_counts()\n",
    "stops_percentage = df['stops'].value_counts(normalize=True) * 100\n",
    "\n",
    "stops_summary = pd.DataFrame({\n",
    "    'Voos': stops_counts,\n",
    "    'Percentual': stops_percentage.round(1)\n",
    "})\n",
    "\n",
    "print(stops_summary)\n",
    "\n",
    "# Insight importante\n",
    "direct_flights = stops_percentage.get('zero', 0)\n",
    "print(f\"\\nüí° INSIGHT: {direct_flights:.1f}% dos voos s√£o diretos (sem paradas)\")\n",
    "print(f\"   Isso sugere oportunidade para mais voos diretos na rota Delhi-Mumbai\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üìä An√°lise Univariada - Vari√°veis Num√©ricas"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Estat√≠sticas descritivas das vari√°veis num√©ricas\n",
    "print(\"üìä ESTAT√çSTICAS DESCRITIVAS - VARI√ÅVEIS NUM√âRICAS\")\n",
    "print(\"=\" * 60)\n",
    "\n",
    "numeric_cols = ['duration', 'days_left', 'price']\n",
    "desc_stats = df[numeric_cols].describe()\n",
    "\n",
    "# Adicionar informa√ß√µes extras\n",
    "extra_stats = pd.DataFrame({\n",
    "    'duration': [df['duration'].skew(), df['duration'].kurt(), df['duration'].mode()[0]],\n",
    "    'days_left': [df['days_left'].skew(), df['days_left'].kurt(), df['days_left'].mode()[0]],\n",
    "    'price': [df['price'].skew(), df['price'].kurt(), df['price'].mode()[0]]\n",
    "}, index=['skewness', 'kurtosis', 'mode'])\n",
    "\n",
    "complete_stats = pd.concat([desc_stats, extra_stats])\n",
    "print(complete_stats.round(2))"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Visualiza√ß√£o das distribui√ß√µes\n",
    "fig, axes = plt.subplots(2, 3, figsize=(18, 12))\n",
    "fig.suptitle('Distribui√ß√µes das Vari√°veis Num√©ricas', fontsize=16, fontweight='bold')\n",
    "\n",
    "# Histogramas\n",
    "for i, col in enumerate(numeric_cols):\n",
    "    # Histograma\n",
    "    axes[0, i].hist(df[col], bins=20, alpha=0.7, color=f'C{i}', edgecolor='black')\n",
    "    axes[0, i].set_title(f'Histograma - {col.title()}')\n",
    "    axes[0, i].set_xlabel(col.title())\n",
    "    axes[0, i].set_ylabel('Frequ√™ncia')\n",
    "    \n",
    "    # Box plot\n",
    "    axes[1, i].boxplot(df[col], patch_artist=True, \n",
    "                       boxprops=dict(facecolor=f'C{i}', alpha=0.7))\n",
    "    axes[1, i].set_title(f'Box Plot - {col.title()}')\n",
    "    axes[1, i].set_ylabel(col.title())\n",
    "\n",
    "plt.tight_layout()\n",
    "plt.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise detalhada dos pre√ßos\n",
    "print(\"üí∞ AN√ÅLISE DETALHADA DOS PRE√áOS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "price_stats = {\n",
    "    'Pre√ßo M√≠nimo': f\"‚Çπ{df['price'].min():,.0f}\",\n",
    "    'Pre√ßo M√°ximo': f\"‚Çπ{df['price'].max():,.0f}\",\n",
    "    'Pre√ßo M√©dio': f\"‚Çπ{df['price'].mean():,.0f}\",\n",
    "    'Pre√ßo Mediano': f\"‚Çπ{df['price'].median():,.0f}\",\n",
    "    'Desvio Padr√£o': f\"‚Çπ{df['price'].std():,.0f}\",\n",
    "    'Amplitude': f\"‚Çπ{(df['price'].max() - df['price'].min()):,.0f}\"\n",
    "}\n",
    "\n",
    "for key, value in price_stats.items():\n",
    "    print(f\"{key:.<20} {value}\")\n",
    "\n",
    "# Quartis e outliers\n",
    "Q1 = df['price'].quantile(0.25)\n",
    "Q3 = df['price'].quantile(0.75)\n",
    "IQR = Q3 - Q1\n",
    "lower_bound = Q1 - 1.5 * IQR\n",
    "upper_bound = Q3 + 1.5 * IQR\n",
    "\n",
    "outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]\n",
    "\n",
    "print(f\"\\nüéØ AN√ÅLISE DE OUTLIERS:\")\n",
    "print(f\"Q1 (25%): ‚Çπ{Q1:,.0f}\")\n",
    "print(f\"Q3 (75%): ‚Çπ{Q3:,.0f}\")\n",
    "print(f\"IQR: ‚Çπ{IQR:,.0f}\")\n",
    "print(f\"Limite Inferior: ‚Çπ{lower_bound:,.0f}\")\n",
    "print(f\"Limite Superior: ‚Çπ{upper_bound:,.0f}\")\n",
    "print(f\"N√∫mero de Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üîó An√°lise Bivariada Inicial"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Pre√ßo m√©dio por companhia a√©rea\n",
    "print(\"üí∞ PRE√áO M√âDIO POR COMPANHIA A√âREA\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "price_by_airline = df.groupby('airline')['price'].agg(['count', 'mean', 'std', 'min', 'max']).round(0)\n",
    "price_by_airline = price_by_airline.sort_values('mean', ascending=False)\n",
    "\n",
    "print(price_by_airline)\n",
    "\n",
    "# Visualiza√ß√£o\n",
    "fig = px.box(df, x='airline', y='price', \n",
    "             title=\"Distribui√ß√£o de Pre√ßos por Companhia A√©rea\",\n",
    "             labels={'price': 'Pre√ßo (‚Çπ)', 'airline': 'Companhia A√©rea'})\n",
    "fig.update_layout(height=500, xaxis_tickangle=-45)\n",
    "fig.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Pre√ßo vs Dura√ß√£o\n",
    "print(\"‚è±Ô∏è RELA√á√ÉO PRE√áO vs DURA√á√ÉO\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "# Correla√ß√£o\n",
    "correlation = df['price'].corr(df['duration'])\n",
    "print(f\"Correla√ß√£o Pre√ßo-Dura√ß√£o: {correlation:.3f}\")\n",
    "\n",
    "# Scatter plot\n",
    "fig = px.scatter(df, x='duration', y='price', \n",
    "                 color='airline', \n",
    "                 size='price',\n",
    "                 title=f\"Rela√ß√£o Pre√ßo vs Dura√ß√£o (Correla√ß√£o: {correlation:.3f})\",\n",
    "                 labels={'duration': 'Dura√ß√£o (horas)', 'price': 'Pre√ßo (‚Çπ)'},\n",
    "                 hover_data=['flight', 'stops'])\n",
    "fig.update_layout(height=600)\n",
    "fig.show()"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# An√°lise por n√∫mero de paradas\n",
    "print(\"üõë PRE√áO M√âDIO POR N√öMERO DE PARADAS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "price_by_stops = df.groupby('stops').agg({\n",
    "    'price': ['count', 'mean', 'std'],\n",
    "    'duration': 'mean'\n",
    "}).round(2)\n",
    "\n",
    "print(price_by_stops)\n",
    "\n",
    "# Insight sobre voos diretos vs com paradas\n",
    "direct_avg = df[df['stops'] == 'zero']['price'].mean()\n",
    "with_stops_avg = df[df['stops'] == 'one']['price'].mean()\n",
    "price_diff = direct_avg - with_stops_avg\n",
    "\n",
    "print(f\"\\nüí° INSIGHT:\")\n",
    "print(f\"Pre√ßo m√©dio voos diretos: ‚Çπ{direct_avg:,.0f}\")\n",
    "print(f\"Pre√ßo m√©dio voos com parada: ‚Çπ{with_stops_avg:,.0f}\")\n",
    "print(f\"Diferen√ßa: ‚Çπ{price_diff:,.0f} ({price_diff/with_stops_avg*100:+.1f}%)\")\n",
    "\n",
    "if price_diff > 0:\n",
    "    print(\"‚ö° Voos diretos s√£o mais caros, como esperado pela conveni√™ncia\")\n",
    "else:\n",
    "    print(\"‚ö° Voos com parada s√£o mais caros - padr√£o inesperado!\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üïê An√°lise Temporal"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Pre√ßo m√©dio por hor√°rio de partida\n",
    "print(\"üïê PRE√áO M√âDIO POR HOR√ÅRIO DE PARTIDA\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "price_by_time = df.groupby('departure_time').agg({\n",
    "    'price': ['count', 'mean', 'std'],\n",
    "    'duration': 'mean'\n",
    "}).round(2)\n",
    "\n",
    "price_by_time_sorted = price_by_time.sort_values(('price', 'mean'), ascending=False)\n",
    "print(price_by_time_sorted)\n",
    "\n",
    "# Visualiza√ß√£o\n",
    "time_price_df = df.groupby('departure_time')['price'].mean().reset_index()\n",
    "fig = px.bar(time_price_df, x='departure_time', y='price',\n",
    "             title=\"Pre√ßo M√©dio por Hor√°rio de Partida\",\n",
    "             labels={'departure_time': 'Hor√°rio de Partida', 'price': 'Pre√ßo M√©dio (‚Çπ)'})\n",
    "fig.update_layout(height=400, xaxis_tickangle=-45)\n",
    "fig.show()"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üéØ Identifica√ß√£o de Padr√µes e Anomalias"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Top 10 voos mais caros\n",
    "print(\"üíé TOP 10 VOOS MAIS CAROS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "top_expensive = df.nlargest(10, 'price')[['airline', 'flight', 'departure_time', 'stops', 'duration', 'price']]\n",
    "print(top_expensive)\n",
    "\n",
    "print(\"\\nüìä Caracter√≠sticas dos voos mais caros:\")\n",
    "expensive_airlines = top_expensive['airline'].value_counts()\n",
    "expensive_times = top_expensive['departure_time'].value_counts()\n",
    "expensive_stops = top_expensive['stops'].value_counts()\n",
    "\n",
    "print(f\"Airlines dominantes: {expensive_airlines.head(3).to_dict()}\")\n",
    "print(f\"Hor√°rios dominantes: {expensive_times.head(3).to_dict()}\")\n",
    "print(f\"Paradas: {expensive_stops.to_dict()}\")"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Top 10 voos mais baratos\n",
    "print(\"üí∞ TOP 10 VOOS MAIS BARATOS\")\n",
    "print(\"=\" * 50)\n",
    "\n",
    "top_cheap = df.nsmallest(10, 'price')[['airline', 'flight', 'departure_time', 'stops', 'duration', 'price']]\n",
    "print(top_cheap)\n",
    "\n",
    "print(\"\\nüìä Caracter√≠sticas dos voos mais baratos:\")\n",
    "cheap_airlines = top_cheap['airline'].value_counts()\n",
    "cheap_times = top_cheap['departure_time'].value_counts()\n",
    "cheap_stops = top_cheap['stops'].value_counts()\n",
    "\n",
    "print(f\"Airlines dominantes: {cheap_airlines.head(3).to_dict()}\")\n",
    "print(f\"Hor√°rios dominantes: {cheap_times.head(3).to_dict()}\")\n",
    "print(f\"Paradas: {cheap_stops.to_dict()}\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üìã Resumo Executivo - Principais Descobertas"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "print(\"üéØ RESUMO EXECUTIVO - PRINCIPAIS DESCOBERTAS\")\n",
    "print(\"=\" * 60)\n",
    "\n",
    "# M√©tricas principais\n",
    "total_flights = len(df)\n",
    "total_airlines = df['airline'].nunique()\n",
    "price_range = (df['price'].min(), df['price'].max())\n",
    "avg_price = df['price'].mean()\n",
    "direct_flights_pct = (df['stops'] == 'zero').mean() * 100\n",
    "\n",
    "# Airlines por faixa de pre√ßo\n",
    "premium_airlines = df.groupby('airline')['price'].mean().nlargest(3)\n",
    "budget_airlines = df.groupby('airline')['price'].mean().nsmallest(3)\n",
    "\n",
    "# Hor√°rios mais caros\n",
    "expensive_times = df.groupby('departure_time')['price'].mean().nlargest(3)\n",
    "\n",
    "print(f\"üìä VIS√ÉO GERAL:\")\n",
    "print(f\"   ‚Ä¢ Total de voos analisados: {total_flights}\")\n",
    "print(f\"   ‚Ä¢ Companhias a√©reas: {total_airlines}\")\n",
    "print(f\"   ‚Ä¢ Faixa de pre√ßos: ‚Çπ{price_range[0]:,} - ‚Çπ{price_range[1]:,}\")\n",
    "print(f\"   ‚Ä¢ Pre√ßo m√©dio: ‚Çπ{avg_price:,.0f}\")\n",
    "print(f\"   ‚Ä¢ Voos diretos: {direct_flights_pct:.1f}%\")\n",
    "\n",
    "print(f\"\\n‚úàÔ∏è SEGMENTA√á√ÉO DE MERCADO:\")\n",
    "print(f\"   ‚Ä¢ Premium: {', '.join(premium_airlines.head(2).index)} (‚Çπ{premium_airlines.iloc[0]:,.0f}+)\")\n",
    "print(f\"   ‚Ä¢ Budget: {', '.join(budget_airlines.head(2).index)} (‚Çπ{budget_airlines.iloc[0]:,.0f}-)\")\n",
    "\n",
    "print(f\"\\nüïê HOR√ÅRIOS PREMIUM:\")\n",
    "for time, price in expensive_times.items():\n",
    "    print(f\"   ‚Ä¢ {time}: ‚Çπ{price:,.0f}\")\n",
    "\n",
    "print(f\"\\nüéØ OPORTUNIDADES IDENTIFICADAS:\")\n",
    "print(f\"   ‚Ä¢ {100-direct_flights_pct:.1f}% dos voos t√™m paradas - oportunidade para voos diretos\")\n",
    "print(f\"   ‚Ä¢ Spread de pre√ßos alto (‚Çπ{price_range[1]-price_range[0]:,}) indica segmenta√ß√£o clara\")\n",
    "print(f\"   ‚Ä¢ Mercado balanceado entre {total_airlines} players principais\")\n",
    "\n",
    "print(f\"\\nüîç PR√ìXIMAS AN√ÅLISES RECOMENDADAS:\")\n",
    "print(f\"   ‚Ä¢ An√°lise estat√≠stica de signific√¢ncia entre grupos\")\n",
    "print(f\"   ‚Ä¢ Modelagem preditiva de pre√ßos\")\n",
    "print(f\"   ‚Ä¢ Segmenta√ß√£o avan√ßada com clustering\")\n",
    "print(f\"   ‚Ä¢ An√°lise de efici√™ncia operacional (pre√ßo/dura√ß√£o)\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## üíæ Exportar Dados Processados"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Criar vers√£o limpa dos dados para pr√≥ximos notebooks\n",
    "import os\n",
    "\n",
    "# Criar diret√≥rio se n√£o existir\n",
    "os.makedirs('../data/processed', exist_ok=True)\n",
    "\n",
    "# Salvar dataset limpo\n",
    "df.to_csv('../data/processed/cleaned_flights_data.csv', index=False)\n",
    "\n",
    # Salvar resumo estat√≠stico
    summary_stats = {
        'dataset_shape': df.shape,
        'missing_values': df.isnull().sum().sum(),
        'duplicates': df.duplicated().sum(),
        'price_stats': df['price'].describe().to_dict(),
        'airlines_count': df['airline'].value_counts().to_dict(),
        'direct_flights_pct': (df['stops'] == 'zero').mean() * 100
    }

    import json
    with open('../data/processed/exploration_summary.json', 'w') as f:
        json.dump(summary_stats, f, indent=2)

    print("‚úÖ Dados exportados com sucesso!")
    print(f"üìÅ Arquivo limpo: ../data/processed/cleaned_flights_data.csv")
    print(f"üìä Resumo estat√≠stico: ../data/processed/exploration_summary.json")

    # Criar arquivo de metadados
    metadata = {
        "data_source": "Airlines Flight Data - Delhi to Mumbai Route",
        "processed_date": pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S'),
        "columns": {
            "airline": "Airline company name",
            "flight": "Flight number/code",
            "departure_time": "Scheduled departure time category",
            "stops": "Number of stops (zero/one)",
            "duration": "Flight duration in hours",
            "days_left": "Days until departure",
            "price": "Flight price in Indian Rupees"
        },
        "key_insights": [
            f"Total flights analyzed: {len(df)}",
            f"Price range: ‚Çπ{df['price'].min():,} - ‚Çπ{df['price'].max():,}",
            f"Average price: ‚Çπ{df['price'].mean():.0f}",
            f"Direct flights percentage: {(df['stops'] == 'zero').mean()*100:.1f}%",
            f"Most expensive airline: {df.groupby('airline')['price'].mean().idxmax()}",
            f"Most budget-friendly airline: {df.groupby('airline')['price'].mean().idxmin()}"
        ]
    }

    with open('../data/processed/data_metadata.json', 'w') as f:
        json.dump(metadata, f, indent=2)

    print(f"üìã Metadados salvos: ../data/processed/data_metadata.json")

# An√°lise de completude dos dados
print("\nüîç AN√ÅLISE DE COMPLETUDE DOS DADOS")
print("=" * 50)

completeness = {}
for col in df.columns:
    non_null = df[col].count()
    total = len(df)
    completeness[col] = {
        'non_null': non_null,
        'null': total - non_null,
        'completeness_pct': (non_null / total) * 100
    }

completeness_df = pd.DataFrame(completeness).T
completeness_df['status'] = completeness_df['completeness_pct'].apply(
    lambda x: '‚úÖ Completo' if x == 100 else f'‚ö†Ô∏è {x:.1f}% completo'
)

print(completeness_df[['non_null', 'null', 'completeness_pct', 'status']])

# Detec√ß√£o de outliers mais robusta
print("\nüéØ DETEC√á√ÉO ROBUSTA DE OUTLIERS")
print("=" * 50)

def detect_outliers_advanced(data, column_name):
    """
    Detecta outliers usando m√∫ltiplos m√©todos
    """
    print(f"\nüìä AN√ÅLISE DE OUTLIERS - {column_name.upper()}")
    print("-" * 40)
    
    # M√©todo IQR
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    iqr_outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    # M√©todo Z-Score (|z| > 3)
    z_scores = np.abs(stats.zscore(data))
    z_outliers = data[z_scores > 3]
    
    # M√©todo Percentil (abaixo de 1% ou acima de 99%)
    p1 = data.quantile(0.01)
    p99 = data.quantile(0.99)
    percentile_outliers = data[(data < p1) | (data > p99)]
    
    print(f"M√©todo IQR: {len(iqr_outliers)} outliers ({len(iqr_outliers)/len(data)*100:.1f}%)")
    print(f"   Limites: [{lower_bound:.0f}, {upper_bound:.0f}]")
    
    print(f"M√©todo Z-Score: {len(z_outliers)} outliers ({len(z_outliers)/len(data)*100:.1f}%)")
    
    print(f"M√©todo Percentil: {len(percentile_outliers)} outliers ({len(percentile_outliers)/len(data)*100:.1f}%)")
    print(f"   Limites: [{p1:.0f}, {p99:.0f}]")
    
    return {
        'iqr_outliers': iqr_outliers,
        'z_outliers': z_outliers,
        'percentile_outliers': percentile_outliers,
        'bounds': {
            'iqr_lower': lower_bound,
            'iqr_upper': upper_bound,
            'percentile_lower': p1,
            'percentile_upper': p99
        }
    }

# Analisar outliers nas vari√°veis principais
outlier_analysis = {}
for col in ['price', 'duration', 'days_left']:
    outlier_analysis[col] = detect_outliers_advanced(df[col], col)

# An√°lise de segmenta√ß√£o natural dos dados
print("\nüìà AN√ÅLISE DE SEGMENTA√á√ÉO NATURAL")
print("=" * 50)

# Segmenta√ß√£o por pre√ßo
price_segments = pd.qcut(df['price'], q=4, labels=['Budget', 'Economy', 'Premium', 'Luxury'])
segment_analysis = df.groupby(price_segments).agg({
    'price': ['count', 'mean', 'std'],
    'duration': 'mean',
    'airline': lambda x: x.mode()[0] if not x.empty else 'N/A'
}).round(2)

print("üè∑Ô∏è SEGMENTA√á√ÉO POR QUARTIS DE PRE√áO:")
print(segment_analysis)

# An√°lise de combina√ß√µes importantes
print("\nüîó AN√ÅLISES DE COMBINA√á√ïES")
print("=" * 50)

# Airline + Stops
airline_stops = pd.crosstab(df['airline'], df['stops'], normalize='index') * 100
print("\n‚úàÔ∏è PERCENTUAL DE VOOS DIRETOS POR AIRLINE:")
print(airline_stops.round(1))

# Departure Time + Price Range
df['price_range'] = pd.qcut(df['price'], q=3, labels=['Baixo', 'M√©dio', 'Alto'])
time_price = pd.crosstab(df['departure_time'], df['price_range'], normalize='index') * 100
print("\nüïê DISTRIBUI√á√ÉO DE PRE√áOS POR HOR√ÅRIO DE PARTIDA:")
print(time_price.round(1))

# M√©tricas de performance de mercado
print("\nüèÜ M√âTRICAS DE PERFORMANCE DE MERCADO")
print("=" * 50)

market_metrics = {}
for airline in df['airline'].unique():
    airline_data = df[df['airline'] == airline]
    
    market_metrics[airline] = {
        'market_share': len(airline_data) / len(df) * 100,
        'avg_price': airline_data['price'].mean(),
        'price_std': airline_data['price'].std(),
        'avg_duration': airline_data['duration'].mean(),
        'direct_flights_pct': (airline_data['stops'] == 'zero').mean() * 100,
        'price_per_hour': (airline_data['price'] / airline_data['duration']).mean()
    }

market_df = pd.DataFrame(market_metrics).T.round(2)
market_df = market_df.sort_values('market_share', ascending=False)

print("üìä DASHBOARD DE M√âTRICAS POR AIRLINE:")
print(market_df)

# Identifica√ß√£o de nichos de mercado
print("\nüéØ IDENTIFICA√á√ÉO DE NICHOS DE MERCADO")
print("=" * 50)

# Voos premium (top 20% por pre√ßo)
premium_threshold = df['price'].quantile(0.8)
premium_flights = df[df['price'] >= premium_threshold]

print(f"‚ú® SEGMENTO PREMIUM (‚â• ‚Çπ{premium_threshold:,.0f}):")
print(f"   ‚Ä¢ {len(premium_flights)} voos ({len(premium_flights)/len(df)*100:.1f}% do mercado)")
print(f"   ‚Ä¢ Airlines dominantes: {premium_flights['airline'].value_counts().head(3).to_dict()}")
print(f"   ‚Ä¢ Hor√°rios preferenciais: {premium_flights['departure_time'].value_counts().head(3).to_dict()}")

# Voos budget (bottom 20% por pre√ßo)
budget_threshold = df['price'].quantile(0.2)
budget_flights = df[df['price'] <= budget_threshold]

print(f"\nüí∞ SEGMENTO BUDGET (‚â§ ‚Çπ{budget_threshold:,.0f}):")
print(f"   ‚Ä¢ {len(budget_flights)} voos ({len(budget_flights)/len(df)*100:.1f}% do mercado)")
print(f"   ‚Ä¢ Airlines dominantes: {budget_flights['airline'].value_counts().head(3).to_dict()}")
print(f"   ‚Ä¢ Caracter√≠sticas: {(budget_flights['stops'] == 'one').mean()*100:.1f}% com paradas")

# Recomenda√ß√µes estrat√©gicas baseadas nos dados
print("\nüí° RECOMENDA√á√ïES ESTRAT√âGICAS")
print("=" * 60)

recommendations = []

# An√°lise de gaps de mercado
direct_vs_stops = df.groupby('stops')['price'].mean()
if direct_vs_stops['zero'] > direct_vs_stops['one']:
    recommendations.append(
        f"üéØ OPORTUNIDADE: Voos diretos s√£o ‚Çπ{direct_vs_stops['zero'] - direct_vs_stops['one']:,.0f} "
        f"mais caros que voos com paradas. H√° demanda por voos diretos budget."
    )

# An√°lise temporal
peak_times = df.groupby('departure_time')['price'].mean().nlargest(2)
recommendations.append(
    f"‚è∞ INSIGHT TEMPORAL: Hor√°rios {list(peak_times.index)} s√£o os mais caros. "
    f"Oportunidade para hor√°rios alternativos."
)

# An√°lise de efici√™ncia
efficiency_by_airline = (df.groupby('airline')['price'] / df.groupby('airline')['duration'].mean()).sort_values()
most_efficient = efficiency_by_airline.index[0]
recommendations.append(
    f"‚ö° EFICI√äNCIA: {most_efficient} oferece melhor custo-benef√≠cio "
    f"(‚Çπ{efficiency_by_airline.iloc[0]:.0f}/hora)"
)

for i, rec in enumerate(recommendations, 1):
    print(f"{i}. {rec}")

print(f"\nüìã RELAT√ìRIO COMPLETO SALVO EM: ../data/processed/")
print(f"   ‚Ä¢ cleaned_flights_data.csv - Dataset limpo")
print(f"   ‚Ä¢ exploration_summary.json - Resumo estat√≠stico")
print(f"   ‚Ä¢ data_metadata.json - Metadados e insights")

print(f"\nüöÄ PR√ìXIMOS PASSOS RECOMENDADOS:")
print(f"   1. Executar an√°lise estat√≠stica inferencial (Notebook 02)")
print(f"   2. Desenvolver modelos preditivos")
print(f"   3. Criar segmenta√ß√£o avan√ßada com clustering")
print(f"   4. Construir dashboard interativo em Streamlit")

# Criar visualiza√ß√£o final resumo
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Distribui√ß√£o de Pre√ßos por Airline',
        'Voos Diretos vs Com Paradas',
        'Pre√ßo vs Dura√ß√£o por Airline',
        'Market Share por Airline'
    ),
    specs=[[{"type": "box"}, {"type": "bar"}],
           [{"type": "scatter"}, {"type": "pie"}]]
)

# Box plot pre√ßos por airline
for airline in df['airline'].unique():
    airline_prices = df[df['airline'] == airline]['price']
    fig.add_trace(
        go.Box(y=airline_prices, name=airline, showlegend=False),
        row=1, col=1
    )

# Bar chart stops
stops_counts = df['stops'].value_counts()
fig.add_trace(
    go.Bar(x=stops_counts.index, y=stops_counts.values, showlegend=False),
    row=1, col=2
)

# Scatter plot price vs duration colored by airline
fig.add_trace(
    go.Scatter(
        x=df['duration'], 
        y=df['price'],
        mode='markers',
        marker=dict(
            color=df['airline'].astype('category').cat.codes,
            colorscale='Set3',
            size=8,
            opacity=0.6
        ),
        text=df['airline'],
        showlegend=False
    ),
    row=2, col=1
)

# Pie chart market share
airline_counts = df['airline'].value_counts()
fig.add_trace(
    go.Pie(
        labels=airline_counts.index, 
        values=airline_counts.values,
        showlegend=False
    ),
    row=2, col=2
)

fig.update_layout(
    height=800,
    title_text="Dashboard Executivo - An√°lise Explorat√≥ria de Dados",
    title_x=0.5
)

fig.show()

print("\n" + "="*60)
print("üéØ EXPLORA√á√ÉO DE DADOS CONCLU√çDA COM SUCESSO!")
print("="*60)
```
